```python

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt  

agg_func={ "asian_pop_15to64":["sum"],
          "black_pop_15to64":["sum"],
          "latino_pop_15to64":["sum"],
          "native_pop_15to64":["sum"],
          "white_pop_15to64":["sum"],
          
          "asian_jail_pop":["sum"],
          "black_jail_pop":["sum"],
          "latino_jail_pop":["sum"],
          "native_jail_pop":["sum"],
          "white_jail_pop":["sum"]}
incarceration=pd.read_csv("D:\LSCE\incarceration_trends-master\incarceration_trends.csv")    
incarceration.fillna(0,inplace=True)
jail_race_year=incarceration.groupby(['year']).agg(agg_func)
jail_race_year.columns = ['_'.join(col).strip() for col in jail_race_year.columns.values]

race_category=["asian","black","latino","native","white"]
for ind in race_category:
    jail_race_year[ind+"_jail_pop_ratio"]=jail_race_year[ind+"_jail_pop_sum"] / jail_race_year[ind+"_pop_15to64_sum"] *100

agg_func={ "asian_pop_15to64":["sum"],
          "black_pop_15to64":["sum"],
          "latino_pop_15to64":["sum"],
          "native_pop_15to64":["sum"],
          "white_pop_15to64":["sum"],
          
          "asian_prison_pop":["sum"],
          "black_prison_pop":["sum"],
          "latino_prison_pop":["sum"],
          "native_prison_pop":["sum"],
          "white_prison_pop":["sum"]
          }
prison_race_year=incarceration.groupby(['year']).agg(agg_func)
prison_race_year.columns = ['_'.join(col).strip() for col in prison_race_year.columns.values]


race_category=["asian","black","latino","native","white"]
for ind in race_category:
    prison_race_year[ind+"_prison_pop_ratio"]=prison_race_year[ind+"_prison_pop_sum"] / prison_race_year[ind+"_pop_15to64_sum"] *100
     
prison_race_ratio=prison_race_year.loc[1990:2015,[s+"_prison_pop_ratio" for s in race_category]]
prison_race_ratio.columns=[s for s in race_category]

jail_race_ratio=jail_race_year.loc[1990:2015,[s+"_jail_pop_ratio" for s in race_category]]
jail_race_ratio.columns=[s for s in race_category]

      
fig = plt.figure(figsize=(6,10))
ax1 = fig.add_subplot(211)
jail_race_ratio.plot(color=['b','g','r','c','m'],ax=ax1)
#plt.rcParams['axes.facecolor'] = 'white'
plt.xlabel("Year",fontsize=15)
plt.ylabel("% jail population",fontsize=15)
plt.title("(a) Change in percent of jail population during the past 25 years",fontsize=15)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc=2, borderaxespad=0.,frameon=False,fontsize=14)

colors=['b','g','r','c','m']
a=-1
for ind in race_category:
    regr = linear_model.LinearRegression()
    regr.fit(np.arange(1990,2016,1).reshape(-1, 1), jail_race_ratio[ind])
    ratio_pred=regr.predict(np.arange(1990,2016,1).reshape(-1,1))
    a+=1
    plt.plot(np.arange(1990,2016,1),ratio_pred,colors[a]+"--")

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.subplots_adjust(wspace =0, hspace =0.5)
ax2 = fig.add_subplot(212)
prison_race_ratio.plot(color=['b','g','r','c','m'],ax=ax2)
plt.xlabel("Year",fontsize=15)
plt.ylabel("% of prison population",fontsize=15)
plt.title("(b) Change in percent of prison population during the past 25 years",fontsize=15)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc=2, borderaxespad=0.,frameon=False,fontsize=14)

colors=['b','g','r','c','m']
a=-1
for ind in race_category:
    regr = linear_model.LinearRegression()
    regr.fit(np.arange(1990,2016,1).reshape(-1, 1), prison_race_ratio[ind])
    ratio_pred=regr.predict(np.arange(1990,2016,1).reshape(-1,1))
    a+=1
    plt.plot(np.arange(1990,2016,1),ratio_pred,colors[a]+"--",alpha=0.6)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#plt.tight_layout()
```

```python
Plot #1: I make this plot to find the change in percent of jail and prison population to total population across different race categories during the 1990-2015. Although the percent of black jail and prison population were well above that of other race categories, there are slight decrease in percent of black jail population and no obvious change in percent of black prison population. Percent of latino jail population decreased significantly. However, the percent of native American jail (prison) population and white jail (prison) population increased significantly during the 1990-2015.
```
